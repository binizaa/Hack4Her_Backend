```python
df=pd.read_csv('merged_limpio.csv')
```

```python
# Diccionario para convertir nombre del mes a número
meses_dict = {
    'Enero': 1, 'Febrero': 2, 'Marzo': 3, 'Abril': 4,
    'Mayo': 5, 'Junio': 6, 'Julio': 7, 'Agosto': 8,
    'Septiembre': 9, 'Octubre': 10, 'Noviembre': 11, 'Diciembre': 12
}

# Crear columna de número de mes
df['Mes_num'] = df['Mes'].map(meses_dict)

# Crear columna de fecha (asignando día 1 para cada mes)
df['Fecha'] = pd.to_datetime(df['Año'].astype(str) + '-' + df['Mes_num'].astype(str) + '-01')

df.drop('Mes', axis=1, inplace=True)
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le_tam = LabelEncoder()
le_cat = LabelEncoder()
le_prod = LabelEncoder()

df["Tamaño de Cliente_encoded"] = le_tam.fit_transform(df["Tamaño de Cliente"])
df["Categoría_encoded"] = le_cat.fit_transform(df["Categoría"])
df["Producto_encoded"] = le_prod.fit_transform(df["Producto"])
```

# 3️⃣ Reto de Volumen — Regresión


Objetivo: Predecir el volumen probable y fijar meta +10% o +20%.

```python
df.head()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder

# 1. Cargar tus datos
df = df

# 2. Seleccionar variables predictoras y target
features = ['Año', 'Mes_num', 'ID Cliente', 'Producto_encoded', 'Tamaño de Cliente_encoded', 'Categoría_encoded']
target = 'Sum of Venta Cajas'

# 3. Dividir entre entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(df[features], df[target], test_size=0.2, random_state=42)

# 4. Modelo de regresión (puedes usar Random Forest o cualquier otro)
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# 5. Predecir volumen en el conjunto original (o uno nuevo para la próxima semana)
df['Volumen Estimado'] = model.predict(df[features])
```

```python
# 6. Producto recomendado: el más comprado históricamente por cliente
producto_mas_comprado = df.groupby('ID Cliente')['Producto_encoded'].agg(lambda x: x.mode().iloc[0])
df_recomendaciones = df[['ID Cliente', 'Volumen Estimado']].drop_duplicates('ID Cliente')
df_recomendaciones = df_recomendaciones.merge(producto_mas_comprado, on='ID Cliente')
df_recomendaciones.rename(columns={'Producto_encoded': 'Producto Recomendado'}, inplace=True)
```

```python
df_recomendaciones
```

```python
# 1️⃣ Aumentar 10% y redondear hacia arriba
df_recomendaciones['Volumen Estimado'] = np.ceil(df_recomendaciones['Volumen Estimado'] * 1.10)

# 2️⃣ Guardar como CSV
df_recomendaciones.to_csv('df_recomendaciones_actualizado.csv', index=False)
```
