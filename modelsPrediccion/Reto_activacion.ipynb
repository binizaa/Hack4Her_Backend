```python
# Diccionario para convertir nombre del mes a número
meses_dict = {
    'Enero': 1, 'Febrero': 2, 'Marzo': 3, 'Abril': 4,
    'Mayo': 5, 'Junio': 6, 'Julio': 7, 'Agosto': 8,
    'Septiembre': 9, 'Octubre': 10, 'Noviembre': 11, 'Diciembre': 12
}

# Crear columna de número de mes
df['Mes_num'] = df['Mes'].map(meses_dict)

# Crear columna de fecha (asignando día 1 para cada mes)
df['Fecha'] = pd.to_datetime(df['Año'].astype(str) + '-' + df['Mes_num'].astype(str) + '-01')

df.drop('Mes', axis=1, inplace=True)
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le_tam = LabelEncoder()
le_cat = LabelEncoder()
le_prod = LabelEncoder()

df["Tamaño de Cliente_encoded"] = le_tam.fit_transform(df["Tamaño de Cliente"])
df["Categoría_encoded"] = le_cat.fit_transform(df["Categoría"])
df["Producto_encoded"] = le_prod.fit_transform(df["Producto"])
```

```python
import pandas as pd

# --- Suponiendo que tu DataFrame se llama df ---

# Asegúrate de que la fecha es tipo datetime
df['Fecha'] = pd.to_datetime(df['Fecha'])

# 1) Encuentra el último mes en la base de datos
ultimo_mes = df['Fecha'].max()

# 2) Crea año-mes para agrupar más fácilmente
df['AñoMes'] = df['Fecha'].dt.to_period('M')

# 3) Agrupa por cliente y año-mes: si tiene ventas, marca 1
ventas_mes = df.groupby(['ID Cliente', 'AñoMes']).agg(
    tiene_venta=('Sum of Venta USD', lambda x: 1 if x.sum() > 0 else 0)
).reset_index()

# 4) Genera una tabla de meses desde el último mes hacia atrás
todos_los_meses = pd.period_range(
    start=ventas_mes['AñoMes'].min(),
    end=ultimo_mes.to_period('M'),
    freq='M'
)

clientes = ventas_mes['ID Cliente'].unique()

# 5) Crea un DataFrame completo de cliente x mes
full = pd.MultiIndex.from_product(
    [clientes, todos_los_meses],
    names=['ID Cliente', 'AñoMes']
).to_frame(index=False)

# 6) Une con ventas reales (para que los meses sin ventas queden como NaN)
full = full.merge(ventas_mes, on=['ID Cliente', 'AñoMes'], how='left')
full['tiene_venta'] = full['tiene_venta'].fillna(0).astype(int)

# 7) Ordena y filtra solo hasta el último mes
full = full[full['AñoMes'] <= ultimo_mes.to_period('M')]
full = full.sort_values(['ID Cliente', 'AñoMes'], ascending=[True, False])

# 8) Calcula racha consecutiva desde el último mes hacia atrás
def contar_racha(grupo):
    racha = 0
    for venta in grupo:
        if venta == 1:
            racha += 1
        else:
            break
    return racha

# Aplica por cliente
racha = full.groupby('ID Cliente')['tiene_venta'].apply(contar_racha).reset_index(name='meses_consecutivos')

# 9) Define categorías
def asignar_categoria(meses):
    if meses >= 12:
        return 'Diamante'
    elif meses >= 9:
        return 'Esmeralda'
    elif meses >= 6:
        return 'Oro'
    elif meses >= 3:
        return 'Plata'
    elif meses >= 1:
        return 'Bronce'
    else:
        return 'Inactivo'

racha['Categoria'] = racha['meses_consecutivos'].apply(asignar_categoria)

# ✅ Resultado: cada cliente con su racha y categoría
print(racha.head())
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

# ------------------------------------------
# 1) Prepara los datos
# ------------------------------------------

# Etiqueta: vamos a predecir si hay COMPRA (>0 cajas) en el siguiente mes
df = df.sort_values(['ID Cliente', 'Fecha'])
df['Compra_siguiente_mes'] = df.groupby('ID Cliente')['Sum of Venta Cajas'].shift(-1)
df['Compra_siguiente_mes'] = (df['Compra_siguiente_mes'] > 0).astype(int)

# Elimina filas sin target (último mes por cliente)
df_modelo = df.dropna(subset=['Compra_siguiente_mes'])

# ------------------------------------------
# 2) Selecciona variables predictoras
# ------------------------------------------

features = [
    'Sum of Venta USD',
    'Sum of Venta Cajas',
    'Tamaño de Cliente_encoded',
    'Categoría_encoded',
    'Producto_encoded',
    'meses_consecutivos'
]

X = df_modelo[features]
y = df_modelo['Compra_siguiente_mes']

# ------------------------------------------
# 3) Divide en train/test
# ------------------------------------------

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# ------------------------------------------
# 4) Entrena modelo
# ------------------------------------------

clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)

# Evalúa
y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred))

# ------------------------------------------
# 5) Genera predicción para clientes inactivos
# ------------------------------------------

# Filtra clientes inactivos en el último mes registrado
inactivos = df[df['Categoria'] == 'Inactivo']

# Opcional: usa solo la última fila por cliente
inactivos = inactivos.sort_values(['ID Cliente', 'Fecha']).groupby('ID Cliente').tail(1)

# Predice probabilidad de compra para cada combinación producto-cantidad hipotética
# Por simplicidad: usa su último producto o el más comprado

# Identifica producto favorito por cliente
producto_fav = (
    df[df['Categoria'] != 'Inactivo']
    .groupby(['ID Cliente', 'Producto_encoded'])
    .agg(total_cajas=('Sum of Venta Cajas', 'sum'))
    .reset_index()
    .sort_values(['ID Cliente', 'total_cajas'], ascending=[True, False])
    .drop_duplicates('ID Cliente')
    [['ID Cliente', 'Producto_encoded']]
)

# Une producto favorito
inactivos = inactivos.merge(producto_fav, on='ID Cliente', how='left', suffixes=('', '_recomendado'))

# Si no tiene historial, usa modo: producto más vendido en general
producto_mas_popular = df.groupby('Producto_encoded')['Sum of Venta Cajas'].sum().idxmax()
inactivos['Producto_encoded_recomendado'].fillna(producto_mas_popular, inplace=True)

# Arma features hipotéticos: supón una cantidad promedio de 1 caja, puedes ajustar esto
inactivos['Sum of Venta USD'] = 0  # precio actual no importa para predicción
inactivos['Sum of Venta Cajas'] = 1  # reto mínimo
inactivos['Producto_encoded'] = inactivos['Producto_encoded_recomendado']

X_inactivos = inactivos[features]

# Predice probabilidad
inactivos['probabilidad_recompra'] = clf.predict_proba(X_inactivos)[:, 1]

# Ordena: mayor probabilidad primero
inactivos = inactivos.sort_values('probabilidad_recompra', ascending=False)

# ------------------------------------------
# 6) Resultado: reto recomendado
# ------------------------------------------

# Output final:
recomendaciones = inactivos[['ID Cliente', 'Producto_encoded', 'Sum of Venta Cajas', 'probabilidad_recompra']]
recomendaciones = recomendaciones.rename(columns={
    'Producto_encoded': 'Producto_recomendado',
    'Sum of Venta Cajas': 'Cantidad_recomendada'
})

print(recomendaciones.head(10))
```
